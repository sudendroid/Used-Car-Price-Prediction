```python
df1 = data_train.copy()
df1 = df1.drop(['New_Price'], axis='columns')
df1.dropna(inplace=True)
df1.shape
```

```python
df1.isna().sum()
```

```python
len(df1.Name.unique())
```

```python
df1['Brand'] = df1['Name'].apply(lambda x: x.split(' ')[0])
df1['Mileage'] = df1['Mileage'].apply(lambda x: x.split(' ')[0])
df1['Power'] = df1['Power'].apply(lambda x: x.split(' ')[0])
df1['Engine'] = df1['Engine'].apply(lambda x: x.split(' ')[0])
df1['Year_Old'] = df1['Year'].apply(lambda x: 2020 - x)
df1.head()
```

From above table it is clear that if we consider VariantName/Name than Fuel_Type, Transmission, Mileage, Engine, Power, Seats, Brand are not usefull because for same variant these values are going to same, so lets drop these fields

RegYear/Year is also useless since we have calculated YearsOld from it, so drop the Year also

```python
df2 = df1.drop(['Fuel_Type', 'Transmission', 'Mileage', 'Engine', 'Power', 'Seats', 'Brand', 'Year'], axis='columns', inplace=False)
df2.head()
```

```python
# Although we can perform OHE for whole dataframe at once but we want to preserver columns names so we will perform them seperately
# dummies = pd.get_dummies(df2)
# dummies.head(3)
```

```python
dummies_name = pd.get_dummies(df2.Name)
dummies_name.head(3)
```

```python
dummies_location = pd.get_dummies(df2.Location)
dummies_location.head(3)
```

```python
dummies_owner_type = pd.get_dummies(df2.Owner_Type)
dummies_owner_type.head(3)
```

```python
df3 = pd.concat([df2.drop(['Name','Location','Owner_Type'], axis='columns'),dummies_name, dummies_location, dummies_owner_type], axis='columns')
df3.head()
```

```python
X = df3.drop('Price', axis='columns')
y = df3.Price
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_reg = LinearRegression()
lr_reg.fit(X_train,y_train)
```

```python
# calculate Mean Absolute error
from sklearn.metrics import mean_absolute_error
y_pred = lr_reg.predict(X_test)
mean_absolute_error(y_test, y_pred)
```

```python
X.head()
```

```python
import numpy as np
def predict_price(name, location, owner_type, km_driven, years_old):    
    loc_index_name = np.where(X.columns==name)[0][0]
    loc_index_location = np.where(X.columns==location)[0][0]
    loc_index_owner = np.where(X.columns==owner_type)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = km_driven
    x[1] = years_old
    if loc_index_name >= 0:
        x[loc_index_name] = 1
    if loc_index_location >= 0:
        x[loc_index_location] = 1
    if loc_index_owner >= 0:
        x[loc_index_owner] = 1

    return lr_reg.predict([x])[0]
```

```python
print(predict_price('Maruti Ertiga VDI','Chennai', 'First', 87000, 7))
print(predict_price('Audi A4 New 2.0 TDI Multitronic','Coimbatore', 'Second', 40000, 7))
print(predict_price('Audi A4 New 2.0 TDI Multitronic','Coimbatore', 'First', 40000, 7))
print(predict_price('Audi A4 New 2.0 TDI Multitronic','Coimbatore', 'First', 10000, 7))
print(predict_price('Audi A4 New 2.0 TDI Multitronic','Coimbatore', 'First', 10000, 2))
```

```python
import pickle
with open('used_car_evaluation_model.pickle','wb') as f:
    pickle.dump(lr_reg,f)
    
print('model saved successfully')
```
